## **Joins - Merge and Concatenation**

```python
df_day = pd.read_csv('/content/day.csv')
df_hour = pd.read_csv('/content/hour.csv')
```

```python
df_hour['dteday']= pd.to_datetime(df_hour['dteday'])  # Convert 'dteday' column to pandas datetime format
df_day['dteday']= pd.to_datetime(df_day['dteday'])  # Convert 'dteday' column to pandas datetime format
```

```python
merged_df = pd.merge(df_hour,df_day[['dteday', 'cnt']], on = 'dteday', how = 'inner')  # Merge df_hour and df_day on 'dteday' column, including 'season' and 'cnt' columns from df_day
merged_df.head()  # Display the first few rows of the merged DataFrame
# merge - The merge() function is used to combine two DataFrames based on a common column or index.
# It allows you to specify how to handle the merging process, such as whether to include all rows from both DataFrames or only matching rows.
```

```python
left_merge = pd.merge(df_hour,df_day[['dteday', 'cnt']], on = 'dteday', how = 'left')
left_merge.shape
```

```python
right_merge = pd.merge(df_hour,df_day[['dteday', 'cnt']], on = 'dteday', how = 'right')
right_merge.head()
```

```python
right_merge.shape
```

```python
outer_merge = pd.merge(df_hour,df_day[['dteday', 'cnt']], on = 'dteday', how = 'outer')
outer_merge.shape # purpose of outer join - it returns all rows from both DataFrames, filling in NaN for missing values where there is no match.
# outer join is used when you want to keep all the data from both DataFrames, even if there are no matching values in the specified column.
```

## **Visualization of Bike Sharing Dataset** Basic Plotting

```python
import matplotlib.pyplot as plt
```

```python
df_day.plot(x='dteday', y='cnt', kind='line',
            figsize=(10, 5), title='Bike Sharing Demand Over Time')  # Plotting the 'cnt' column against 'dteday' column
plt.xlabel('Date')  # Setting the x-axis label
plt.ylabel('Total_rentals')  # Setting the y-axis label
plt.grid
plt.show()  # Display the plot
```

```python
# shows the total number of rentals over time, allowing us to visualize the trend and seasonality in bike sharing demand.
```

```python
df_day.plot(x='dteday', y='cnt', kind='line',
            figsize=(10, 5), title='Bike Sharing Demand Over Time',marker = 'o',linestyle = 'dotted')  # Plotting the 'cnt' column against 'dteday' column
plt.xlabel('Date')  # Setting the x-axis label
plt.ylabel('Total_rentals')  # Setting the y-axis label
plt.grid
plt.show()  # Display the plot
```

```python
df_day['cnt'].hist(bins = 20, color = 'skyblue', edgecolor = 'black',figsize = (8, 5))  # Plotting a histogram of the 'cnt' column with 20 bins
plt.title('Distribution of Total Rentals')  # Setting the title of the histogram
plt.xlabel('Rentals Count')  # Setting the x-axis label
plt.ylabel('Frequency')  # Setting the y-axis label
plt.grid(True)
plt.show()  # Display the histogram
# The histogram shows the distribution of total rentals,
# allowing us to see the frequency of different rental counts and identify any patterns or trends in the data.
# The histogram provides insights into the distribution of bike rentals, helping to identify peaks and troughs in demand.
# The histogram shows the distribution of total rentals,
# allowing us to see the frequency of different rental counts and identify any patterns or trends in the data.
# It the data is skewed or not, and whether there are any outliers or unusual values in the data.
```

```python
df_day.boxplot(column='cnt', grid=False, figsize=(10, 5))  # Creating a box plot for the 'cnt' column
plt.title='Box Plot of Total Bike Rentals'
plt.show()
```

```python
# helps me to find extremes, outliers, and the spread of the data.
# boxplot has 1st quartile, 2nd quartile, 3rd quartile, minimum value, and maximum value.
```

```python
df_season = df_day.groupby('season')['cnt'].mean()

df_season.plot(kind='line',
               marker='o',
               color='red',
               figsize=(8, 5),
               title = 'Average Rentals by season')  # Plotting a bar chart of average rentals by season
plt.xlabel('Season(1: Spring, 2: Summer, 3: Fall, 4: Winter)')  # Setting the x-axis label
plt.ylabel('Average Rentals')  # Setting the y-axis label
plt.grid(True)
plt.show()
```

```python
df_season.plot(kind='bar',
               color=['blue','orange','green','red'],
               figsize=(8, 5),
               title = 'Average Rentals by season')  # Plotting a bar chart of average rentals by season
plt.xlabel('Season')  # Setting the x-axis label
plt.ylabel('Total Rentals')  # Setting the y-axis label
plt.xticks(rotation=0)  # Rotating x-axis labels for better readability
plt.show()
```
